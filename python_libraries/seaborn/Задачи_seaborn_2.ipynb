```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
sns.set()
```

```python
df = pd.read_csv("titanic.csv", sep=",")
df.head(5)
```

```python
sns.countplot(x="class", hue="sex", data=df, palette="Set1")
```

```python
sns.countplot(x="deck", hue="sex", data=df, palette="Set1")
```

```python
sns.barplot(x="deck", y="survived", hue="class", data=df, palette="Set1", estimator=sum, errorbar=None)
```
